```python
import pandas as pd
```

```python
data = {
    'idade': [18,74,52,35,19,45],
    'salario': [1500,5600,4800,3500,1800,2750],
    'comprou': [0,1,1,0,0,1]
}

df = pd.DataFrame(data)
df
```

```python
x = df[['idade','salario']]
# x = df.drop('comprou', axis=1)
y = df['comprou']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
modelo = LogisticRegression()
modelo.coef_
```

```python
modelo.fit(x,y)
modelo.coef_
```

```python
data_teste = {
      'idade': [38,52],
      'salario': [3500,4800]
}
df_teste = pd.DataFrame(data_teste)
df_teste
```

```python
previsao = modelo.predict(df_teste)
previsao
```

```python
data = {
    "gasto": [100, 150, 200 ,250, 300, 350],
    "compras": [20,30,40,50,60,70]
}
df = pd.DataFrame(data)
df
```

```python
y = df['compras']
x = df[['gasto']]
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
modelo.fit(x,y)
```

```python
df_teste = pd.DataFrame({
    "gasto": [1000]
})

previsao = modelo.predict(df_teste)
previsao
```
